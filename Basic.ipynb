```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Scores = pd.DataFrame({'RollNo':[1,2,3,4],'Marks':[23,34,49,44]})
Scores
```

```python
sns.lineplot(x='RollNo',y='Marks',data =Scores)
plt.title('Student Marks')
```

```python
sns.get_dataset_names()
```

```python
crashes = sns.load_dataset('car_crashes')
Tips = sns.load_dataset('tips')
crashes.head()
```

```python
sns.scatterplot(x='total',y='not_distracted',data =crashes,hue ='alcohol',size ='speeding',palette='YlGnBu')
```

```python
sns.histplot(Tips['tip'], kde=True, bins=30)
```

```python
sns.displot(Tips['tip'],kde =True,bins = 15)
```

```python
sns.barplot(x='sex',y='tip',data =Tips,palette ='YlGnBu')
```

```python
sns.boxplot(x='day',y='tip',data =Tips,hue='sex',palette ='YlGnBu')
```

```python
sns.stripplot(x='day',y='tip',data =Tips,hue='sex',palette ='YlGnBu',dodge=True)
```

```python
sns.jointplot(x='tip',y='total_bill',data =Tips,kind ='hex')
```
